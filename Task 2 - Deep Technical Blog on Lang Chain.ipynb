```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate
```

```python

template_1 = PromptTemplate(
    input_variables=["topic"],
    template="""
Explain the topic "{topic}" in simple and easy words.

- Start with a short definition
- Use simple language
- Give one real-life example
- Keep it short and clear
"""
)

print(template_1.format(topic="Artificial Intelligence"))
```

```python
template_2 = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="""
Explain the topic "{topic}" for {audience} in a {tone} tone.

- Start with a simple definition
- Use language suitable for {audience}
- Keep the tone {tone}
- Give one easy example
- Keep the explanation short and clear
"""
)

print(template_2.format(topic="AI", audience="beginners", tone="friendly"))
print(template_2.format(topic="Python", audience="kids", tone="fun"))
print(template_2.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

```python
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="""
Explain "{topic}" step by step in a simple way.

- Start with a short definition
- Break it into clear steps
- Use simple language
- Give one real-life example
"""
)

# Interview Template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="""
Generate 3 interview questions about "{topic}".

- Keep questions clear and relevant
- Mix easy and moderate level
"""
)

# Story Template
story_template = PromptTemplate(
    input_variables=["topic"],
    template="""
Explain "{topic}" as a short and simple story.

- Use easy language
- Make it engaging
- Relate it to real life
"""
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

```python


# System message (defines role clearly
system_template = SystemMessagePromptTemplate.from_template(
    """
You are a helpful {role}.

- Explain concepts clearly
- Use simple language
- Give examples when possible
"""
)

# Human message (better instruction)
human_template = HumanMessagePromptTemplate.from_template(
    """
Explain "{topic}" in a simple way.

- Start with a short definition
- Break it into easy points
- Give one real-life example
"""
)

# Combine into chat prompt
chat_prompt = ChatPromptTemplate.from_messages([
    system_template,
    human_template
])

# Input
messages = chat_prompt.format_messages(
    role="teacher",
    topic="Neural Networks"
)

# Output messages
for msg in messages:
    print(msg)
```

```python
def validate_inputs(audience, tone):
    valid_audience = {"beginner", "intermediate", "expert"}
    valid_tone = {"formal", "casual", "fun"}

    # Normalize input (handle case issues like "Beginner", "FUN")
    audience = audience.lower()
    tone = tone.lower()

    if audience not in valid_audience:
        print(f"Invalid audience '{audience}', defaulting to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print(f"Invalid tone '{tone}', defaulting to 'formal'")
        tone = "formal"

    return audience, tone


print(validate_inputs("beginner", "fun"))
print(validate_inputs("invalid", "invalid"))
```

```python
def generate_prompt(topic, audience, tone, style):
    # Validate inputs
    audience, tone = validate_inputs(audience, tone)

    # Normalize style input
    style = style.lower()

    styles = {
        "teaching": """
Explain "{topic}" for {audience} in a {tone} teaching style.

- Start with a simple definition
- Break into steps
- Give one example
""",
        "interview": """
Generate 3 interview questions about "{topic}" for {audience} in a {tone} tone.

- Include easy to moderate questions
""",
        "storytelling": """
Explain "{topic}" for {audience} in a {tone} storytelling style.

- Make it engaging
- Use simple language
- Relate to real life
"""
    }

    # Handle invalid style safely
    if style not in styles:
        print(f"Invalid style '{style}', defaulting to 'teaching'")
        style = "teaching"

    template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=styles[style]
    )

    return template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )


# Example
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```
